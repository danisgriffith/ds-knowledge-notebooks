## Introduction

<br>
A <b>Multi Layer Perceptron (MLP) is a class of feed-forward artificial neural network consisting of at least three layers of nodes</b>: the input layer, many potential hidden layers, and the output layer.

<br>
<br>

    <img src="images/multi_layer_perceptron.png" alt="forward propagation" width="70%" height="70%">
</a>

<br>
As we briefly mentioned in the first paragraph, MLP contain multiple neurons (nodes) arranged in layers; let's dive a little into the details of this structure :

<br>
<ul style="list-style-type:square">
    <li>
        the <b>input layer</b> represents all of the input variables; no computation is performed in any of the input nodes,
        they just pass on the information to the first hidden layer
    </li>
    <br>
    <li>
        the <b>hidden layer(s)</b> collect the weighted signals from all the input nodes and perform additional computation, 
        essentially creating a new set of features for each hidden layer. This new set of signals will then serve as the input
        vector for either the next hidden layer of the output layer. When working with more than one hidden layer, the
        architecture is called "deep neural network", otherwise it is said to be "shallow".
    </li>
    <br>
    <li>
        the <b>output layer</b> collects the signals from the last hidden layer and perform an ultimate level of computation
        which yields (depending on the task assigned) a prediction or a classification
    </li>
</ul>

<br>
All nodes in each layer are connected to all the nodes in the next layer (except for the output nodes, of course). Graphically, we visualize these connections as lines, which represent the weighted contribution of one node to a node in the next layer or, equivalently, the flow of information from one node to the next.

<br>
It is worth pointing out that <b>in a feed-forward neural network (thus both SLP and MLP) the information flows in only one direction : from the input nodes, through the hidden nodes (if any) and to the output nodes</b>. Other types of neural networks have more intricate connections, such as feedback paths.


## Representational Power

<br>
One intriguing property of Multi Layer Perceptron is its <b>ability to discover useful intermediate representations at the hidden unit layers inside the network</b> : because training examples constrain only the network inputs and outputs, the weight-tuning procedure is free to set weights that define whatever hidden unit representation is most effective at minimizing the value of the cost function. <b>This can lead the algorithm to define new hidden layer features that are not explicit in the input representation, but which capture properties of the input instances that are most relevant to learning the target function</b>.

<br>
This ability of multilayer networks to automatically discover useful representations at the hidden layers is a key feature of ANN learning. <b>In contrast to learning methods that are constrained to use only predefined features provided by the human designer, this provides an important degree of flexibility that allows the learner to invent features not explicitly introduced by the human designer</b>. Note when more layers of units are used in the network, more complex features can be invented.

The ability of single layer perceptrons to represent AND, OR, NAND, and NOR remains nonetheless important because every boolean function can be represented by some network of interconnected units based on these primitives. In fact, every boolean function can be represented by some network of perceptrons only two levels deep, in which the inputs are fed to multiple units, and the outputs of these units are then input to a second, final stage.

<br>
Because networks of threshold units can represent a rich variety of functions and because single units alone cannot, we are generally interested in learning multi layer networks of threshold units.
