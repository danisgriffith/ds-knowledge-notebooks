# VIOLATION OF LINEARITY

<br>


## Introduction

<br>
From the notebook about clrm assumptions, recall that the assumption of linearity (<b>A1</b>), also known as the assumption on the functional form, states that the population regression equation takes the form 

<br>
<blockquote>
$ 
    \quad
    \mathbf{Y} = \mathbf{X} \ \boldsymbol{\beta} + \boldsymbol{\varepsilon} 
    \quad \text{or} \quad
    \mathbf{Y}_i 
        = \mathbf{E}[\mathbf{Y}_i \mid \mathbf{X}_i] + {\boldsymbol{\varepsilon}_i}
        = \boldsymbol{\mathbf{X}_i}^{\top} \boldsymbol{\beta} + {\boldsymbol{\varepsilon}_i}
$
</blockquote>

<br>
<blockquote>
    The expected value of the response variable is assumed to be a linear (with respect to the parameters) combination of the
    independent variables; the latter can take non-linear forms (or be arbitrarily transformed, as in the case of
    polynomial regression) as long as the parameters are linear.
</blockquote>


## Decomposition of the estimated response variable

<br>
Let's start by bringing back the two equations concerned with functional form : 

<br>
<blockquote>
$
    \quad
    \begin{align}
        \mathbf{Y} \ = \ \mathbf{X} \ \boldsymbol{\beta} + \boldsymbol{\varepsilon}
        & \quad \textbf{PRE}
        \newline
        \hat{\mathbf{Y}}
        \ = \ \mathbf{X} \ \boldsymbol{\hat{\beta}} + \boldsymbol{\mathbf{e}}
        & \quad \textbf{SRE}
    \end{align}
$
</blockquote>

<br><br>
The Polulation Regression Equation (PRE) states that the dependent variable consists of two terms : 

<br>
<ul style="list-style-type:square">
    <li>
        the Population Regression Function (PRF, also known as the systematic component) $ \ f(\mathbf{X}) \ $,
        which describes the relationship between the dependent and the independent variables as a linear (with respect to the
        unknown population parameters) function of the independent variables     
    </li>
    <br>
    <li>
        the disturbance term $\boldsymbol{\varepsilon}$ (also known as error term or stochastic component), which accounts for
        the deviation of the observed value $\boldsymbol{\mathbf{Y}}$ from its expected value $ f(\mathbf{X}) $
    </li>
</ul>

<br><br>
By analogy, we can think of the estimated response variable $\hat{\mathbf{Y}}$ as the result of two components :

<br>
<ul style="list-style-type:square">
    <li>
        the <b>deterministic portion</b>, represented by the Sample Regression Function (SRF); this is the part of
        $\hat{\mathbf{Y}}$ explained by the independent variables. All of the explanatory (predictive) information of the model
        should be in this portion  
    </li>
    <br>
    <li>
        the <b>stochastic portion</b>, represented by the residuals $\boldsymbol{e}$ 
    </li>
</ul>

<br>
The idea is that the deterministic portion of our model is so good at explaining (or predicting) the dependent variable that only the inherent randomness of a real-world phenomenon remains leftover for the stochastic term. 

<br>
If we observe explanatory or predictive power in the stochastic term, we know that the deterministic component is missing some of the predictive information.


## Residuals-vs-Fits plot


<br>
When conducting a residual analysis, a "residuals versus fits plot" is the most frequently created plot. It is a scatter plot of the residuals on the vertical axis and the fitted values on the horizontal axis.

<br>
By mean of the residual plot we can assess whether the residuals are consistent with the hypothesis we made about the stochastic component of the estimated response variable. Residual plots are also used to detect non-linearity, unequal error variances, and outliers.

<br>
First we will take a look at the example below, to see what a "well-behaved" residual plot looks like, and then we will analyze.


```python
import linear_regression as lr
lr.plot_residuals_vs_fits(cmd = 'linearity_correct', inline = True)
```

<br>
The first plot suggests that there is a decreasing linear relationship between the regressor and the dependent variable.

<br>
It also illustrates that there is no unusual data point in the data set, and that the variation around the estimated regression line is approximately constant, but these considerations are out of the scope of the current investigation. In the previous sentence the word "unusual" refers to outliers, which is a different concept from influential points.

<br>
The second plot is the actual residuals-vs-fits plot. It's easy to see that the <b>horizontal line</b> is the locus of points (where each point here is a residual) describing a perfect match between the observed and the estimated values of the response variable; in other words, a residual that is found on this horizontal line corresponds to an observed value on the <b>estimated regression line</b> in the first plot.

<br>
The residuals should not be either systematically high or low, and thus be <c>centered on zero</c> throughout the range of fitted values. This is equivalent to say that the model is correct on average for all fitted values. In the OLS context random errors are also assumed to produce residuals that are <b>normally distributed</b>; therefore the latter should fall in a <b>symmetrical pattern</b> and have a <b>constant spread</b> throughout the range.

<br>
<font color='red'>
When the residuals "bounce randomly" around the zero horizontal line, it's reasonable to think that the assumption of linearity was justified. 

<br>
<font color='red'>
Here are the characteristics of a well-behaved residual-vs-fits plot and what they suggest about the appropriateness of the simple linear regression model :

<br>
<ul style="list-style-type:square">
    <li>
         the residuals "bounce randomly" around the 0 line, this suggests that the assumption that the relationship is linear is
         reasonable
    </li>
    <br>
    <li>
        the residuals roughly form a sort of "horizontal band" around the zero horizontal line; this suggests that the error
        terms have a constant variance, regardless of the values of the regressors (out of the scope of assessing linearity)
    </li>
    <br>
    <li>
         no residual "stands out" from the random pattern, this suggests that there are no outliers (out of the scope of
         assessing linearity)
    </li>
</ul>

<br><font color='black'>
Now that we know what a "correct" residuals-vs-fits plot should look like, let's ask ourselves : <br>
how does a non-linear functional form show up on this type of plot ?


```python
lr.plot_residuals_vs_fits(cmd = 'linearity_violation')
```

<br>
Even though the first plot may suggest the existence of a linear relationship between the regressor and the dependent variable, it's clear from the residuals-vs-fits plot that this is not the case. In the example above, the residuals depart from the zero horizontal line to follow a quadratic function.

<br>
Any <b>systematic (non-random) pattern</b> is sufficient to say that the relationship is not linear.

<br>
In multiple regression models, nonlinearity or nonadditivity may also be revealed by systematic patterns in plots of the residuals vs individual independent variables.

<br>
<font color='red'>
Incidentally, did you notice that the r2 value is very high (95.26%)? This is an excellent example of the caution "a large r2 value should not be interpreted as meaning that the estimated regression line fits the data well." The large r2 value tells you that if you wanted to predict groove depth, you'd be better off taking into account mileage than not. The residuals vs. fits plot tells you, though, that your prediction would be better if you formulated a non-linear model rather than a linear one.

## Correction TODO

<br>
Even though models are sometimes non-linear in the parameters, some may still be intrinsically linear. In such circumstances, the problem of non-linearity can be solved by first transforming the data in such away that it becomes linear, likely in regards to a new set of parameters; for example, by applying logarithms or some other function to the appropriate independent or dependent variables.

<br>
Another possible solution to the problem of non-linearities is to apply <b>transformations</b> to the independent variables (prior to fitting) which map these variables into a higher dimension space. If the transformations are chosen properly, then even if the original data is not well modeled by a linear function, the transformed data will be. To automate such a procedure, the Kernel Principle Component Analysis technique and other so called Nonlinear Dimensionality Reduction techniques can automatically transform the original data into a new feature space that is chosen to capture important characteristics of the data.

<br>
$
    \quad
    \begin{align}
        \mathbf{Y} &= 
                      \boldsymbol{\beta_0} \ \boldsymbol{\mathbf{X}_0}  
                    + \boldsymbol{\beta_1} \ \boldsymbol{\mathbf{X}_1}  
                    + \dots 
                    + \boldsymbol{\beta_{p-1}} \ \boldsymbol{\mathbf{X}_{p-1}}  
                    + \boldsymbol{\varepsilon} 
        \newline
        \boldsymbol{\mathbf{Y}^*} &= 
                      \boldsymbol{\theta_0} \ \boldsymbol{ \mathbf{X}_0 }  
                    + \boldsymbol{\theta_1} \ \boldsymbol{ (\mathbf{X}_1)^3 }  
                    + \dots 
                    + \boldsymbol{\theta_{p-1}} \ \boldsymbol{ (\mathbf{X}_{p-1})^2 }   
                    + \boldsymbol{\varepsilon}     
    \end{align}
$

<br>
Another method for avoiding non-linearity is to apply a <b>non-parametric regression method such as local linear regression or a kernel regression technique</b>. These non-parametric algorithms usually involve setting a model parameter (such as a smoothing constant for local linear regression or a bandwidth constant for kernel regression) which can be estimated using a technique like cross validation. Both these approaches can model very complicated systems, but the performance can rapidly erode as the dimensional space of the independent variables increase.

<br>
Another solution is to directly apply <b>"kernel method" like support vector regression or kernelized ridge regression</b> with an appropriate choice of a non-linear kernel function. These methods automatically apply linear regression in a non-linearly transformed version of your feature space (with the actual transformation used determined by the choice of kernel function) which produces non-linear models in the original feature space. 

<br>
An important idea to be aware of is that it is typically better to apply a method that will automatically determine how much complexity can be afforded when fitting a set of training data than to apply an overly simplistic linear model that always uses the same level of complexity. 

<br>
When the support vector regression technique and ridge regression technique use linear kernel functions (and hence are performing a type of linear regression) they generally avoid overfitting by automatically tuning their own levels of complexity, but even so cannot generally avoid underfitting (since linear models just are not complex enough to model some systems accurately when given a fixed set of features). The kernelized versions of these techniques, however, can avoid both overfitting and underfitting since they are not restricted to a simplistic linear model.
